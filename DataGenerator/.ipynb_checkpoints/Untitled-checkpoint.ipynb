```python
from species_generator import generate_species
from name_generator import generate_names
import pandas as pd
import random
```

```python
frame_columns = ["PatientName","Attendings","IPCStaff","Pathogen"] 
patient_info = pd.DataFrame(columns=frame_columns) #init a blank table
num_patients = 100 
patient_names = generate_names(100,"f l")
attendings_names = generate_names(30,"f m l")
ipc_names = generate_names(20,"f m l")
for patient in patient_names:
    curr_attending = random.choice(attendings_names)
    curr_staff = ";".join(random.choices(ipc_names,k=random.randint(1,5)))
    curr_pathogen = ";".join(generate_species(random.randint(1,3),bool(random.randint(1,1))))
    row = [patient,curr_attending,curr_staff,curr_pathogen]
    patient_info = patient_info.append(pd.DataFrame([row],columns=frame_columns),ignore_index = True)
```
